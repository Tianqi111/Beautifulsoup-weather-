```python
# import the libraries 
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
url = 'https://www.hko.gov.hk/en/cis/normal/1981_2010/dnormal10.htm'
page = requests.get(url)
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#get information that we need
date = soup.find_all('td', headers= 'date')
press = soup.find_all('td', headers= 'press')
temp_tmax = soup.find_all('td', headers= 'temp tmax')
temp_tmean = soup.find_all('td', headers= 'temp tmean')
temp_tmin = soup.find_all('td', headers= 'temp tmin')
wet = soup.find_all('td', headers= 'wet')
dew = soup.find_all('td', headers= 'dew')
rh = soup.find_all('td', headers= 'rh')
rain = soup.find_all('td', headers= 'rain')
cloud = soup.find_all('td', headers= 'cloud')
```

```python
#get text of the information we need
date_l = [n.get_text() for n in date]
press_l = [n.get_text() for n in press]
temp_tmax_l = [n.get_text() for n in temp_tmax]
temp_tmean_l = [n.get_text() for n in temp_tmean]
temp_tmin_l = [n.get_text() for n in temp_tmin]
wet_l = [n.get_text() for n in wet]
dew_l = [n.get_text() for n in dew]
rh_l = [n.get_text() for n in rh] 
rain_l = [n.get_text() for n in rain]
cloud_l =[n.get_text() for n in cloud]
```

```python
#the length of the list should be same, clean the useless data in "Date" and "Press"
date_l.pop(-1)
press_l.pop(-1)
```

```python
# build the csv
hk_Observatory = pd.DataFrame({
    "Date": date_l,
    "Mean Pressure (hPa) Figure" : press_l,
    "Mean Maximum (deg.C)" : temp_tmax_l,
    "Mean (deg.C)" : temp_tmean_l,
    "Mean Minimum (deg.C)" : temp_tmin_l,
    "Wet Bulb (deg.C) Figure" : wet_l,
    "Dew Point (deg.C) Figure" : dew_l,
    "Pelative Humidity (%) Figure" : rh_l,
    "Mean Daily Rainfall (mm) Figure" : rain_l,
    "Amount of Cloud (%) Figure" : cloud_l,
})
hk_Observatory
```

```python
#change the index into Date
hk_Observatory.set_index("Date")
```

```python
#save the csv document
hk_Observatory.to_csv('./Data/hk_Observatory.csv')
```
